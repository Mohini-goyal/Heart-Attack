# Splitting in x and y variable

```python
x = df.drop("output", axis=1)
y = df[["output"]]
```

# train test split

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.20, random_state= 101)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

# Scaling appling on the x data

```python
st = StandardScaler()
x_sc_tr = st.fit_transform(x_train)
x_sc_test = st.transform(x_test)
```

```python
x_sc_tr
```

# Model

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(x_sc_tr,y_train)
```

# Prediction

```python
train_pre_lg = lg.predict(x_sc_tr)
test_pre_lg = lg.predict(x_sc_test)
```

# Accuracy

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
acc_tr_lg = accuracy_score(y_train, train_pre_lg)
acc_test_lg = accuracy_score(y_test, test_pre_lg)
print("acc_test_lg -", acc_test_lg)
print("acc_tr_lg - ", acc_tr_lg)
print("**"*10)
confusion_tr_lg = confusion_matrix(y_train, train_pre_lg)
confusion_test_lg = confusion_matrix(y_test, test_pre_lg)
print("confusion_test_lg - \n", confusion_test_lg)
print("confusion_tr_lg - \n", confusion_tr_lg)
print("**"*10)
class_tr_lg = classification_report(y_train, train_pre_lg)
class_test_lg = classification_report(y_test, test_pre_lg)
print("class_test_lg - \n", class_test_lg)
print("class_tr_lg - \n", class_tr_lg)
```

# Decision Tree

# Prediction

# Accuracy

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
acc_tr_dt = accuracy_score(train_pre_dt, y_train)
acc_test_dt = accuracy_score(test_pre_dt, y_test)
print("acc_test_dt -", acc_test_dt)
print("acc_tr_dt - ", acc_tr_dt)
print("**"*10)
confusion_tr_dt = confusion_matrix(train_pre_dt, y_train)
confusion_test_dt = confusion_matrix(test_pre_dt, y_test)
print("confusion_test_dt - \n", confusion_test_dt)
print("confusion_tr_dt - \n", confusion_tr_dt)
print("**"*10)
class_tr_dt = classification_report(train_pre_dt, y_train)
class_test_dt = classification_report(test_pre_dt, y_test)
print("class_test_dt - \n", class_test_dt)
print("class_tr_dt - \n", class_tr_dt)
```
